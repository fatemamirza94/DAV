```

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
dfPM2_5ao = pd.DataFrame(dfao, columns=['PM2.5','PM10','NO2','CO','SO2','O3']   )#
correlation = dfPM2_5ao.corr()
print("\n")
print("Pearson Standard Correlation Coefficient Matrix in Aotizhongxin")
print(correlation)
print("\n")
 
f, ax = plt.subplots(figsize=(11, 9))
ax.set_title('Correlation map between air pollutants in Aotizhongxin')
sns.set_style("white")
cmap = sns.diverging_palette(230, 20, as_cmap=True)
 
heatmap = sns.heatmap(correlation, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},
            annot = True, annot_kws = {"size": 12})
 
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})

correlation.to_csv('corr_ao.csv')
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
AAPL_price = pd.read_csv('/content/Aotizhongxin.csv',usecols=['date', 'PM2.5'])
SPY_price = pd.read_csv('/content/Aotizhongxin.csv',usecols=['date', 'PM10'])
X = sm.add_constant(SPY_price['PM10'])
model = sm.OLS(AAPL_price['PM2.5'],X)
results = model.fit()
plt.scatter(SPY_price['PM10'],AAPL_price['PM2.5'],alpha=0.3)
y_predict = results.params[0] + results.params[1]*SPY_price['PM10']
plt.plot(SPY_price['PM10'],y_predict, linewidth=3)
plt.xlim(240,350)
plt.ylim(100,350)
plt.xlabel('SPY_price')
plt.ylabel('AAPL_price')
plt.title('OLS Regression')
print(results.summary())
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
AAPL_price = pd.read_csv('/content/Aotizhongxin.csv',usecols=['date', 'PM10'])
SPY_price = pd.read_csv('/content/Aotizhongxin.csv',usecols=['date', 'PM2.5'])
X = sm.add_constant(SPY_price['PM2.5'])
model = sm.OLS(AAPL_price['PM10'],X)
results = model.fit()
plt.scatter(SPY_price['PM2.5'],AAPL_price['PM10'],alpha=0.3)
y_predict = results.params[0] + results.params[1]*SPY_price['PM2.5']
plt.plot(SPY_price['PM2.5'],y_predict, linewidth=3,color = 'red')
#plt.xlim(240,350)
#plt.ylim(100,350)
plt.xlabel('SPY_price')
plt.ylabel('AAPL_price')
plt.title('OLS Regression')
print(results.summary())
```
